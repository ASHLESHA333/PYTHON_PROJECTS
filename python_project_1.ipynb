```python
## filling null values
for i in df.columns:
    if df[i].dtypes==object:
        df[i]=df[i].fillna(df[i].mode()[0]) ## FOR OBJECT VALUES
    else:
        df[i]=df[i].fillna(df[i].mean()) ## FOR NUMERICAL VALUES
```

```python
##  Which batsman has scored the highest number of runs?

total_runs_by_batsman = df.groupby('batsman')['batsman_runs'].sum()
sorted_total_runs = total_runs_by_batsman.sort_values(ascending=False)
sorted_total = total_runs_by_batsman.sort_values(ascending=False).head(1)
print(f"The batsman with the highest number of runs is {sorted_total}")

```

```python
## What is the average number of extras (wide runs, bye runs, legbye runs, noball runs, penalty runs) conceded per inning by each bowling team?

df['total_extras'] = df['wide_runs'] + df['bye_runs'] + df['legbye_runs'] + df['noball_runs'] + df['penalty_runs']
extras_inning = df.groupby(['bowling_team', 'inning'])['total_extras'].sum()
average_inning = extras_inning.groupby('bowling_team').mean()
print("Average number of extras conceded per inning by each bowling team:")
print(average_inning)

```

```python
plt.figure(figsize=(12, 6))
average_inning.plot(kind='bar', color='pink')
plt.title('Average Extras Conceded Per Inning by Each Bowling Team')
plt.xlabel('Bowling Team')
plt.ylabel('Average Extras')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
